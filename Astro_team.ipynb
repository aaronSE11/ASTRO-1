```python
#Este archivo lo paso Aaron en el whatsapp
data = pd.read_excel("plantas solares mexico.xlsx")
```

```python
data
```

```python
#Eliminamos las columnas que no aportan nada
data.drop([1,3,5,7], inplace = True)
```

```python
#Aqui cambiamos las columnas por renglones, ya que es mas facil manipular columnas
data = data.T
data.columns = ['Centrales','Estado','Municipio','Capacidad','Coordenadas']
```

```python
data
```

```python
#No se pueden leer las coordenadas con el formato de vector, asi que agrego dos columnas mas para la latitud y longitud
data[['lat', 'lon']] = data['Coordenadas'].str.split(', ', expand=True)
```

```python
data
```

```python
#Originalmente el formato para la info es tipo object, asi que lo pasamos a algo que si se pueda operar
data['lat'] = data['lat'].astype(float)
data['lon'] = data['lon'].astype(float)
data['Centrales'] = data['Centrales'].astype(str)
```

```python
data.head(5)
```

```python
#Esta funcion genera el mapa en bruto, con una localizacion inicial en el defacult
def generatemap(default_location=[32.391562,-115.270062], default_zoom_start=6):
	m = folium.Map(location=[32.391562,-115.270062], zoom_start=4,min_zoom=3, max_bounds=True)
	return m
m = generatemap()


#Aqui se agregan los marcadores azules en cada coordenada del dataframe
for i in tqdm(range(0,len(data))):
   folium.Marker(
      location=[data.iloc[i]['lon'], data.iloc[i]['lat']],
      popup=data.iloc[i]['Centrales'],
   ).add_to(m)
```

```python
m
```

```python
#Esto es solo por si quieren guardar el mapa como un html, pero les 
#recomiendo comentarlo. Solo lo puse para enviarselos por whatsapp
m.save('mapa_paneles_mexico.html')
```
